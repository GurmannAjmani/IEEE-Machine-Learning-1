```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tvmarketing.csv")
```

```python
def loss(m,c,points):
    t_err=0
    for i in range(len(points)):
        x=points.iloc[i].TV
        y=points.iloc[i].Sales
        t_err+=((y-(m*x+c))**2)/float(len(points))
```

```python
def gradient_descent(m_now,c_now,points,lr):
    m_gradient=0
    c_gradient=0
    n=len(points)
    for i in range(n):
        x=points.iloc[i].TV
        y=points.iloc[i].Sales
        m_gradient+= -(2/n) * x * (y-(m_now * x + c_now))
        c_gradient+= -(2/n) *(y-(m_now * x + c_now))
    m=m_now - m_gradient * lr
    c=c_now - m_gradient * lr
    return m,c
```

```python
m=0
c=0
lr=0.000005
iterations=1000
for i in range(iterations):
    m,c=gradient_descent(m,c,df,lr)
```

```python
fig,ax=plt.subplots(figsize=(10,7))
plt.xlabel("TV")
plt.ylabel("Sales")
plt.scatter(df.TV,df.Sales,c="red")
plt.plot(list(range(0,300)),[m*x+c for x in range(0,300)],color="black")
```
